```python
class BankAccount():
    def __init__(self, owner):
        self.owner = owner
        self.balance_file = ("./%s.balance.txt" %owner)
        self.history_file = ("./%s.history.txt" %owner)
        
    def get_balance(self):
        with open(self.balance_file, 'r') as f:
            self.balance = int(f.read())
        return self.balance
    
    def set_balance(self, amount):
        with open(self.balance_file, 'w') as f:
            f.write("%d" %amount)
    
    def execute_deposit(self, amount, user):
        bal = self.get_balance()
        bal += amount
        self.set_balance(bal)
        
user1_account = BankAccount("user1")
user1_account.execute_deposit(1000, "user2")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def milk_graph():
    data = pd.read_csv('milk.csv')
    data = data.set_index('year')
    grouped_dairy = data.groupby('dairy')
    
    for key, group in grouped_dairy:
        group_sorted = group.sort_values('year', ascending=True)
        group_sorted.plot(kind='bar', subplots=False)
        plt.legend(['production', 'consumption'])
        plt.xlabel(key)    
        plt.show()
```

```python
milk_graph()
```
