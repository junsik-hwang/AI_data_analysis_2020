```python
# 2.3 데이터 슬라이싱
# (1) 열(Column) = 변수 = Feature = Label
diamonds.loc[:5, ["carat", "price"]]
```

```python
# (2) 행(row) : 행의 위치를 알 때, 조건을 만족하는 행
diamonds.loc[diamonds.carat > 4]
```

```python
diamonds.loc[(diamonds.carat > 4) & (diamonds.cut == 'Fair'), ["x", "y", "z"]]
```

```python
diamonds.iloc[0:10,[7,8,9]]
```

```python
# 2.4 새로운 변수 만들기
diamonds["xyz_mean"] = (diamonds.x + diamonds.y + diamonds.z) / 3
```

```python
diamonds.head()
```

```python
# 2.5 변환 : Box-Cox Transformation
# (1) log 변환
diamonds["price_log"] = np.log10(diamonds.price)
diamonds.head()
```

```python
# (2) root 변환
diamonds["price_root"] = np.sqrt(diamonds.price)
diamonds.head()
```

```python
# (3) inverse 변환
diamonds["price_inverse"] = 1 / diamonds.price
diamonds.head()
```

```python
# 양적 자료(수치형 자료) -> 질적 자료(범주형 자료)
# carat -> carat_group
# 0 ~ 2 : "light"
# 2 ~ 4 : "Middle"
# 4 ~ 6 : "Heavy"
diamonds["carat_group"] = pd.cut(diamonds.carat,
                                bins   = [0, 2, 4, 6],
                                right  = False,
                                labels = ["Light", "Middle", "Heavy"])
diamonds.tail()
```

```python
# 열 삭제하기
# (1) del data["variable"]
# (2) data.drop(["variable1, variable2"], axis = 1, inplace = True)

del diamonds["xyz_mean"]
diamonds.head()
```

```python
diamonds.drop(["price_log", "price_root", "price_inverse", "carat_group"], 
              axis    = 1, 
              inplace = True)
diamonds.head()
```

```python
# 2.6 데이터 정렬하기
diamonds.sort_values(by = "carat", ascending = True).head()
```

```python
diamonds.sort_values(by = ["carat", "price"], ascending = [False, True]).head()
```

```python
# 2.7 데이터 합치기
# (1) merge : inner join, outer join(full join, left join, right join)
d1 = pd.DataFrame({"id" : [1, 2, 3, 4],
                  "bt" : ["a", "b", "ab", "o"]})

d2 = pd.DataFrame({"id" : [3, 4, 5, 6],
                  "height" : [160, 170, 180, 190],
                  "weight" : [60, 70, 80, 90]})

print(d1)
print(d2)
```

```python
# inner join : 두 데이터프레임의 primary key가 동일한 데이터만 합쳐짐

d3 = pd.merge(d1, d2, on = "id", how = "inner")
d3
```

```python
# outer join: 두 데이터프레임을 모조리 합친다. (값이 없는 곳은 NaN)

d4 = pd.merge(d1, d2, on = "id", how = "outer")
d4
```

```python
# left join : merge에 먼저 들어가는 d1이 모두 들어오고 이에 따른 d2 데이터가 들어온다. (right join은 그 반대)
d5 = pd.merge(d1, d2, on = "id", how = "left")
d5
```
