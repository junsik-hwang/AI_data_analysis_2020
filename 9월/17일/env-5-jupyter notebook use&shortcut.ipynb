# Notebook user interface

# Modal editor

## Edit mode

- green cell border
- 선택한 셀에 입력이 가능한 상태
- `커서가 깜빡임`
- `Enter`로 모드 전환

## Command mode


- grey cell border with a blue left margin
- 선택한 셀에 입력이 불가능한 상태
- `커서 깜빡임이 없음`
- Edit->Command mode전환은 `ESC`

# Running Code

- `Shift-Enter`: runs the current cell and move below
- `Alt-Enter`:  runs the current cell and inserts a new one below
- `Ctrl-Enter`: run the current cell and enters command mode

# Cell 관련 단축키

`단축키 도움말: ( Command Mode ) + H`

## Cell 생성하기

`현재 셀 위에(Above) 셀 만들기 : ( Command Mode ) + A`  
`현재 셀 아래(Below) 셀 만들기 : ( Command Mode ) + B`

## Cell의 형식을 변경

1. Code 셀 : 코드를 입력하기 위한 셀
2. Markdown 셀 : 코드가 아닌 텍스트를 입력하기 위한 셀, 마크다운 서식 사용 가능

`Code 셀로 변경 :      ( Command Mode ) + Y`  
`Markdown 셀로 변경 :  ( Command Mode ) + M`
