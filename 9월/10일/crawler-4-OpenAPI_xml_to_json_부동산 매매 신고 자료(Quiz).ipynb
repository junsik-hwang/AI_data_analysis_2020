## 최대 개수가 있는 index 얻기
- 가장 뒤쪽이 최대 개수가 있는 index임
- max_n일 때만 idx_list에 저장함

````python
max_n = 0
idx_list = []
for idx,item in enumerate(estate_json['response']['body']['items']['item']):
    if max_n < len(item):
        max_n = len(item)
    elif max_n == len(item):
        idx_list.append(idx)
    else:
        pass
````
