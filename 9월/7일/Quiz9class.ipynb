```python
class Item():
    def __init__(self, name, count, expire_day, notice):
        self.name = name
        self.count = count
        self.expire_day = expire_day
        self.notice = notice
        
    def start():
        print("---- 재고 관리 프로그램 ----")
        print("  1 - 현재 재고 상태 출력")
        print("  2 - 재고 수정")
        print("  3 - 검색(품목명)")
        print("  4 - 검색(유통기한)")
        print("  9 - 종료")
        return int(input("명령번호 : "))
        
    def show_count(self):
        goods = list(items.keys())
        goods_cnt = list(items.values())
        goods_exp = list(expire.values())
        goods_not = list(notice.values())
        print()
        print("----- 재고 수량 -----")
        for i in range(len(items)):
            print(f"{goods[i]} : 재고 수량({goods_cnt[i]}), 유통기한({goods_exp[i]}), 주의사항({goods_not[i]})")
    def update_count(self):
        print()
        print("----- 재고 수정 -----")
        for _ in list(items.keys()):
            print(_, end = ' ')
        print()
        goods = input("품목명 : ")
        goods_cnt = int(input("수량 : "))
        if(goods in items.keys()):
            items[goods] = goods_cnt
            print(f"{goods} 품목 수량이 {goods_cnt}개로 수정되었습니다.")
            print(end = '\n')
        else:
            print(f"{goods} 품목은 존재하지 않습니다.")
    def search_goods(self):
        print()
        print("----- 검색 : 품목명 -----")
        goods = input("검색 할 품목명 : ")
        print(f"{goods} : 수량({items[goods]}), 유통기한({expire[goods]}), 주의사항({notice[goods]})")
        print()
    
    def search_expire(self):
        print()
        print("----- 검색 : 유통기한 -----")
        exp = int(input("유통기한 : "))
        print(f"유통기한이 {exp}일 이상인 품목 : ", end = '')
        for _ in list(expire.keys()):
            if(expire[_] >= exp):
                print(_, end = ' ')
        print()
        print()
        

goods_box = Item(["coffee", 0, 7, "냉장보관"], ["pen", 0, 0, "없음"])
num = 0
while(num != 9):
    num = goods_box.start()
    if(num == 1):
        show_count()
    elif(num == 2):
        update_count()
    elif(num == 3):
        search_count()
    elif(num == 4):
        search_goods()
    elif(num == 9):
        break
    else:
        print("올바르지 않은 명령어입니다.")
```

```python
class Item():
    def __init__(self, name, count, expire_day, notice):
        self.name = name
        self.count = count
        self.expire_day = expire_day
        self.notice = notice
        
    def start(self):
        print("---- 재고 관리 프로그램 ----")
        print("  1 - 현재 재고 상태 출력")
        print("  2 - 재고 수정")
        print("  3 - 검색(품목명)")
        print("  4 - 검색(유통기한)")
        print("  9 - 종료")
        return int(input("명령번호 : "))
    
    def show_count(self):
        goods = self.name
        goods_cnt = self.count
        goods_exp = self.expire_day
        goods_not = self.notice
        print()
        print("----- 재고 수량 -----")
        for i in range(len(goods)):
            print(f"{goods[i]} : 재고 수량({goods_cnt[i]}), 유통기한({goods_exp[i]}), 주의사항({goods_not[i]})")
            
    def update_count(self):
        print()
        print("----- 재고 수정 -----")
        for _ in self.name:
            print(_, end = ' ')
        print()
        goods_name = input("품목명 : ")
        goods_count = int(input("수량 : "))
        if(goods_name in self.name):
            self.count[self.name.index(goods_name)] = goods_count
            print(f"{goods_name} 품목 수량이 {goods_count}개로 수정되었습니다.")
            print(end = '\n')
        else:
            print(f"{goods_name} 품목은 존재하지 않습니다.")
            
    def search_goods(self):
        print()
        print("----- 검색 : 품목명 -----")
        goods = input("검색 할 품목명 : ")
        goods_index = self.name.index(goods)
        print(f"{goods} : 수량({self.count[goods_index]}), 유통기한({self.expire_day[goods_index]}), 주의사항({self.notice[goods_index]})")
        print()
    
    def search_expire(self):
        print()
        print("----- 검색 : 유통기한 -----")
        exp = int(input("유통기한 : "))
        print(f"유통기한이 {exp}일 이상인 품목 : ", end = '')
        for a,b in enumerate(self.expire_day):
            if(b >= exp):
                print(self.name[a], end = ' ')
        print()
        print()
    
goods_box = Item(["coffee","pen", "apple", "milk", "coke", "book"], [0, 0, 0, 0, 0, 0], [7, 0, 20, 5, 30, 0], ["냉장보관","없음", "상온보관", "냉장보관", "흔들지 말것", "없음"])
num = 0
while(num != 9):
    num = goods_box.start()
    if(num == 1):
        goods_box.show_count()
    elif(num == 2):
        goods_box.update_count()
    elif(num == 3):
        goods_box.search_goods()
    elif(num == 4):
        goods_box.search_expire()
    elif(num == 9):
        break
    else:
        print("올바르지 않은 명령어입니다.")
```
